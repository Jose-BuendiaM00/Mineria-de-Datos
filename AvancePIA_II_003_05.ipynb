```python
import requests
import plotly.express as px
df=pd.read_csv('TOT_MEX.csv')
repo_url = 'https://raw.githubusercontent.com/angelnmara/geojson/master/mexicoHigh.json' #Archivo GeoJSON
mx_regions_geo = requests.get(repo_url).json()

fig = px.choropleth(data_frame=df, 
                    geojson=mx_regions_geo, 
                    locations='Estado', # nombre de la columna del Dataframe
                    featureidkey='properties.name',  # ruta al campo del archivo GeoJSON con el que se hará la relación (nombre de los estados)
                    color='Numero', #El color depende de las cantidades
                    color_continuous_scale="deep", #greens
                   )
fig.update_geos(showcountries=True, showcoastlines=True, showland=True, fitbounds="locations")

fig.update_layout(
    title_text = 'Casos de atentados en los estados de México',
    font=dict(
        #family="Courier New, monospace",
        family="Ubuntu",
        size=18,
        color="#7f7f7f"
    ),
    annotations = [dict(
        x=0.55,
        y=-0.1,
        xref='paper',
        yref='paper',
        text='Fuente: <a href="https://www.kaggle.com/START-UMD/gtd">\
            Kaggle </a>',
        showarrow = False
    )]
)

fig.show()

#Referencia: https://youtu.be/MaypM_PUuPk
```

Es notable que en los estados del Centro y del Sur la incidencia es mayor comparada a los estados del norte. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

datos=pd.read_csv('TOT_MEX.csv')
df=pd.DataFrame(datos)
df.groupby('Estado')['Numero'].sum().plot(kind='bar',legend='Reverse')
plt.xlabel('Total de ataques por Estado')

#Referencia: https://youtu.be/-f7BxfSC6Es
```

Esta grafica apoya el mapa anterior y nos ayuda a tener una mejor idea, ya que representa a todos los estados de los que se tienen registro y considerar a los estados con mayor numero de casos (**CDMX, Chiapas, Guerrero, Jalisco, Oaxaca**).

```python
ar=pd.read_csv('COMP_MEX.csv',encoding='utf8')
nuevo=ar[['country_txt','iyear','attacktype1','targtype1']]
sns.set(style='ticks',color_codes=True)
g=sns.pairplot(nuevo,hue='country_txt',palette='tab10')

#Referencia: https://youtu.be/-f7BxfSC6Es
```

* Se puede observar que a lo largo de los años, los ciudadanos o lugares o grupos más susceptibles a recibir ataques son: Negocios, Politicos, Gobierno (General), Policia y Periodistas. 

* Así mismo, podemos pensar que los negocios frecuentemente reciben asaltos a mano armada, o tambien pueden recibir ataques con bombas, o igual a edificios del gobierno pueden recibir ataques explosivos, otro ataque tentativo que se puede observar es el secuestro, tanto como a Policias, Politicos, Periodistas y dueños de negocios. 

```python
em=pd.read_csv('COMP_MEX5.csv',encoding='utf8')
nuevo=em[['provstate','iyear','attacktype1','weaptype1','targtype1']]
sns.set(style='ticks',color_codes=True)
g=sns.pairplot(nuevo,hue='provstate',palette='tab10')

#Referencia: https://youtu.be/-f7BxfSC6Es
```
